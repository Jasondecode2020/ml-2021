- 15 complex operation to a column, df_copy['Temperature'].apply(handle_temp)

```python
def handle_temp(x):
    if x > 20:
        return 20
    else:
        return x
```

```python
df_copy['Temperature'] = df_copy['Temperature'].apply(handle_temp)
```

```python
df_copy['Temperature']
```

- 18 unique and groupby and agg, df['CloudCover'].unique()

```python
df['CloudCover'].unique()
```

```python
group = df.groupby('CloudCover').agg({'Temperature':['mean']})
```

```python
group
```

```python
group.columns = group.columns.droplevel(0)
group.rename(columns={ group.columns[0]: 'Temp_mean' }, inplace = True)
```

```python
group.head()
```

- 19 merge, pd.merge(df, group, how='left', on='CloudCover')

```python
df = pd.merge(df, group, how='left', on='CloudCover')
df
```

- 20 stack

```python
part1 = df.iloc[:10] # 10 rows
part2 = df.iloc[20:] # last rows
data_stacked = part1.append(part2, ignore_index=True) # stack together
```

```python
data_stacked
```
